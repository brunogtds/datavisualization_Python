```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd

%matplotlib inline 
```

```python
x= np.arange(-10,10,0.01)

plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.sin(4*x), label='sin(4*x)')
plt.title('Exemplo de sin(4*x) e cos(x)')
plt.xlabel('eixo X')
plt.ylabel('eixo y')
plt.legend()
plt.show()
```

```python
###script layer -> plt
###artist layer -> axis e fig
```

```python
plt.figure(figsize=(12,4))
plt.suptitle('Script layer', fontsize=14, weight='bold', y=1.05)

plt.subplot(1,2,1) #linhas, colunas, gr√°fico ativo
plt.plot(x, np.sin(x))
plt.title('Plot sin(x)')

plt.subplot(1,2,2)
plt.plot(x, np.cos(4*x))
plt.title('Plot cos(4*x)')

plt.tight_layout() #Adjust the padding between and around subplots.
plt.show()
```

```python
fig, ax= plt.subplots(2,2,figsize=(12,4))
plt.suptitle('Artist layer', fontsize=14, weight='bold', y=1.05)

ax[0][0].plot(x, np.sin(x))
ax[0][0].set_title('Sin (x)')

ax[0][1].plot(x, np.sin(5*x))
ax[0][1].set_title('Sin 5*x')

ax[1][0].plot(x, np.cos(x))
ax[1][0].set_title('Cos (x)')

ax[1][1].plot(x, np.cos(10*x))
ax[1][1].set_title('Cos (10*x)')

plt.tight_layout() #Adjust the padding between and around subplots.
plt.show()
```

```python
x = np.arange(0, 6, 0.3)
y = np.arange(0, 6, 0.3)

fig= plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title('Explicit and short  parameters', fontsize=14, weight='bold')
plt.plot(x, np.sin(x), color='springgreen', marker='o', linestyle='dashed', linewidth=2,  markersize=4)
plt.plot(y, np.cos(y), color='slateblue', linestyle='dashdot', linewidth=2)
plt.plot(x, np.tanh(x), '--r', linewidth=2, markersize=6)
plt.plot(x, 1/np.exp(x), '-.sb', linewidth=1, markersize=4)

plt.subplot(1,2,2)
plt.title('Enfatizing lines', fontsize=14, weight='bold')
plt.plot(x, np.cos(x), linewidth=1, alpha=0.7)
plt.plot(x, np.sin(x), linewidth=1, alpha=0.7)
plt.plot(x, np.tanh(x), linewidth=1, alpha=0.7)
plt.plot(x, 1/np.exp(x), '-.ok')

plt.tight_layout() #Adjust the padding between and around subplots.
plt.show()
```

```python
x= np.arange(0.0, 1.1, 0.01)

fig= plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
plt.title('Estilos de linha', fontsize=14, weight='bold')
plt.plot(x, x**1, linestyle='solid', label='solid')
plt.plot(x, x**2, linestyle='dashed', label='dashed')
plt.plot(x, x**3, linestyle='dashdot', label='dashdot')
plt.plot(x, x**4, linestyle='dotted', label='dotted')
plt.legend(title='Estilos de linha')

plt.subplot(1,3,2)
plt.title('N√≠vel de cinza', fontsize=14, weight='bold')
plt.plot(x, x**1, color='1.0', label='1.0')
plt.plot(x, x**2, color='0.75', label='0,75')
plt.plot(x, x**3, color='0.5', label='0,5')
plt.plot(x, x**4, color='0.25', label='0,25')
plt.plot(x, x**5, color='0.1', label='0,1')
plt.legend(title='N√≠vel cinza')

plt.subplot(1,3,3)
plt.title('Markers', fontsize=14, weight='bold')
plt.plot(x, x**1, marker='o', label='o')
plt.plot(x, x**2, marker='x', label='x')
plt.plot(x, x**3, marker='v', label='v')
plt.plot(x, x**4, marker='>', label='>')
plt.plot(x, x**6, marker='s', label='s')
plt.plot(x, x**7, marker='D', label='D')
plt.legend(title='Markers')


plt.tight_layout() 
plt.show()
```

```python
x= np.linspace(-3,3,20)

fig= plt.figure(figsize=(20,8))

plt.suptitle('fill between', fontsize=14, weight='bold', y=1.05)

plt.subplot(2,2,1)
plt.plot(x, np.cos(x))
plt.fill_between(x, np.cos(x), -1, alpha=0.3, color='g')
plt.text(-0.6,-0.2,'Sob a curva',color='red',fontsize=5)
plt.ylim([-1.2,1.2])

#plt.text()

plt.subplot(2,2,2)
plt.plot(x, np.tanh(x))
plt.fill_between(x, np.tanh(x), alpha=0.3)
plt.text(-2.4, -0.5, 'tanh(x) < 0', color='red', fontsize=5)
plt.text(1.2, 0.4, 'tanh(x) > 0', color='red', fontsize=5)
plt.hlines(0,-3,3, linestyle='dashed', color='b', linewidth=1)

from scipy.stats import norm

pdf=[]
cdf=[]

for a in np.arange(-4,4,0.05):
  pdf.append(norm.pdf(a))
  cdf.append(norm.cdf(a))

plt.subplot(2,2,3)
plt.title('PDF, &N(\mu=0, \sigma=1)$', fontsize=12)
plt.plot(np.arange(-4,4,0.05),pdf)

plt.fill_between(np.arange(-4,4,0.05), pdf, alpha=0.3)
plt.fill_between(np.arange(-4,-0.5,0.05), pdf[0:len(np.arange(-4,-0.5,0.05)+1)], color='g', alpha=0.6)
plt.text(-1.4, 0.06, f'√Årea = { norm.cdf(-0.5) :.3f}',fontsize=5,color='red')

plt.subplot(2,2,4)
plt.title('CDF, SN(\mu=0, \sigma=1$',fontsize=12)
plt.plot(np.arange(-4,4,0.05),cdf)

plt.fill_between(np.arange(-4,4,0.05),cdf,alpha=0.3)
plt.fill_between(np.arange(-4,-0.5,0.05),cdf[0:len(np.arange(-4,-0.5,0.05)+1)],color='g',alpha=0.6)

plt.vlines(-0.5,0,max(cdf)-max(cdf)/2,linewidth=2,linestyle=':',colors='g')
plt.hlines(norm.cdf(-0.5),-4,0,linewidth=2,linestyle=':',colors='g')
plt.plot(-0.5,norm.cdf(-0.5),'go')
plt.text(0.2, norm.cdf(-0.5), f'P(x<-0.5) = {norm.cdf(-0.5) :.3f}',fontsize=12,color='white')



plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as ticker

fig, ax= plt.subplots(1,2,figsize=(12,4)) #sharey=True,sharex=True
plt.suptitle('Configurando os eixos', fontsize=12, weight='bold')

x= np.linspace(-2,2,2000)

ax[0].set_title('$cos(x_i)$ e $tan(x_i)$', fontsize=14)
ax[0].plot(x, np.cos(x))
ax[0].plot(x, np.tanh(x))

ax[0].xaxis.set_major_locator(ticker.MultipleLocator(0.75))
ax[0].yaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax[0].grid()

ax[1].set_title('$cos(x_i)$ e $tan(x_i)$', fontsize=14) ###latex
ax[1].plot(x, np.cos(x))
ax[1].plot(x, np.tanh(x))

ax[1].xaxis.set_major_locator(ticker.NullLocator())
ax[1].yaxis.set_major_locator(ticker.MaxNLocator(8))
ax[1].grid()

plt.tight_layout()
plt.show()
```

```python
x= np.random.random_sample(100)
np.random.seed(1234)

x= np.arange(0,100,1)
y= np.random.rand(100)*10
z= np.random.rand(100)*3
t= np.random.rand(100)*5

from matplotlib import cm #cores no scatterplot

plt.figure(figsize=(12,12))
plt.suptitle('V√°rios tipos de gr√°ficos', fontsize=16, weight='bold', y=1.05)

plt.subplot(3,2,1)
plt.plot(x,y)
plt.title('Plot', fontsize=12, weight='bold')

plt.subplot(3,2,2)
plt.hist(y, bins=30, color='g', alpha=0.5)
plt.title('Histograma', fontsize=12, weight='bold')

plt.subplot(3,2,3)
plt.scatter(x,y,c=x,cmap=cm.viridis)
plt.title('Scatter plot', fontsize=12, weight='bold')

plt.subplot(3,2,4)
plt.boxplot([y,z,t],labels=['Cyan','Green','Yellow'])
plt.title('Box plot', fontsize=12, weight='bold')

plt.subplot(3,2,5)
plt.bar(['Cyan','Green','Yellow'],[y.mean(),z.mean(),t.mean()],color=['c','g','y'],alpha=0.5)
plt.title('Bar plot', fontsize=12, weight='bold')

plt.subplot(3,2,6)
plt.pie([y.mean(),z.mean(),t.mean()], colors=['c','g','y'],explode=[0, 0.1, 0], labels=['Cyan','Green','Yellow'],autopct='%1.2f%%')
plt.title('Pie plot', fontsize=12, weight='bold')

plt.tight_layout()
plt.show()

```
