```python
dias=tips.day
dias=dias.to_list()
dias.count('Fri')
```

```python
selected_tips=tips[['day','tip','size']] ##retorna um novo dataframe
selected_tips
```

```python
dolar_real_rate= 5.07
tips['total_bill'] = tips['total_bill'] * dolar_real_rate
tips['tip'] = tips['tip'] * dolar_real_rate
tips
```

```python
tips_fumantes = tips[tips.smoker == 'Yes']
tips_nao_fumantes = tips[tips.smoker != 'Yes']
```

```python
tips_nao_fumantes
```

```python
tips[tips.smoker=='Yes']['tip'].mean() > tips[tips.smoker=='No']['tip'].mean()
```
